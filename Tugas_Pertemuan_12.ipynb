```python
# Buatlah program dengan input dan fungsi untuk menghitung:
  # a. Luas persegi panjang
  # b. Volume persegi panjang
  # c. Luas lingkaran
  # d. Volume bola

# a. Luas persegi panjang (pxl)

def luas_persegi_panjang (panjang, lebar):
  hasil = panjang * lebar
  return hasil
print("===Luas persegi panjang===")

panjang = int(input("Masukkan panjangnya: "))
lebar = int(input("Masukkan lebarnya: "))
luas = panjang * lebar

print(("Luas persegi panjangnya adalah"), luas)

# b. Volume persegi panjang (pxlxt)

def Volume_persegi_panjang (panjang, lebar, tinggi):
  hasil = panjang * lebar * tinggi
  return hasil
print("\n===Volume persegi panjang===")

panjang = int(input("Masukkan panjangnya: "))
lebar = int(input("Masukkan lebarnya: "))
tinggi = int(input("Masukkan tingginya: "))
luas = panjang * lebar * tinggi

print(("Volume persegi panjangnya adalah"), luas)

# c. Luas lingkaran (3.14xr**2)

def luas_lingkaran(r):
  hasil = 3.14 * r * r
  return hasil

print("\n===Luas lingkaran===")

r = int(input("Masukkan jari-jari lingkarannya: "))
luas = int(luas_lingkaran(r)) #untuk menghilangkan bilangan setelah koma

print("Luas lingkarannya adalah:", luas)

# d. Volume bola (4/3x3.14xr**3)

def volume_bola(r):
  hasil = (4/3) * 3.14 * r**3
  return hasil
print("\n===Volume bola===")

r = int(input("Masukkan jari jarinya: "))
volume = int(volume_bola(r))

print("Volume bolanya adalah:", volume)



```
