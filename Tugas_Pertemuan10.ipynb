```python
# 1. Tuples:
# a. Buat sebuah tuple bernama mahasiswa yang berisi data diri anda dengan 3 nilai:
# (“Nama”, “NIM”,”Prodi”)
# b. Ubah prodi menjadi “Teknologi Informasi dan Elektro”
# c. Tampilkan data NIM dan Prodi

# a. Buat sebuah tuple bernama mahasiswa yang berisi data diri anda dengan 3 nilai: (“Nama”, “NIM”,”Prodi”)
mahasiswa = ("Fatkhan Maulana Atmaja", "2505060053", "Hukum")
print(f"Nama: {mahasiswa[0]}")
# b. Ubah prodi menjadi “Teknologi Informasi dan Elektro”
mahasiswa_update = list(mahasiswa)
# Mengganti "Hukum" ke "Teknologi Informasi dan Elektro"
mahasiswa_update[2] = "Teknologi Informasi dan Elektro"
# Diubah kembali menjadi tuple
mahasiswa = tuple(mahasiswa_update)

# c. Tampilkan data NIM dan Prodi
print(f"NIM: {mahasiswa[1]}")
print(f"Prodi: {mahasiswa[2]}")
```

```python
#a. Buat dictionary keranjang dengan data: {"apel": 5000, "pisang": 3000, "jeruk": 7000,"anggur": 15000}
keranjang = {
    "apel": 5000,
    "pisang": 3000,
    "jeruk": 7000,
    "anggur": 15000,
}
total_harga = 0
# b. Gunakan perulangan untuk menghitung total harga semua barang
for barang, harga in keranjang.items():
    print(f"Harga {barang}: Rp {harga}")
    total_harga += harga
# d. Tampilkan harga total belanja
print(f"\nHarga total belanja: Rp {total_harga}")
```
